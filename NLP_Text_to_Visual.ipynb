```python
import networkx as nx
import matplotlib.pyplot as plt
from adjustText import adjust_text

# Initialize the graph
G = nx.Graph()

# Define the concepts and their frequencies
concepts = {
    "Peace": 15,
    "Community Building": 6,
    "Compassion and Empathy": 3,
    "Role of Education": 4,
    "Inclusion": 3,
    "Cultural Diversity": 8,
    "Social Responsibility": 3,
    "Role of Youth": 12,
    "Leadership": 2,
    "Dreams and Aspirations": 7,
    "Self-love": 5,
    "Personal Identity": 5,
    "Mental Health Awareness": 2,
    "Overcoming Adversity": 4,
    "Resilience": 3,
    "Environmental Awareness": 2,
    "Global Citizenship": 2,
    "Empathy": 3
}

# Add nodes with sizes proportional to their frequencies
for concept, size in concepts.items():
    G.add_node(concept, size=size * 100)

# Define the relationships (edges) between concepts
edges = [
    ("Peace", "Community Building"),
    ("Peace", "Compassion and Empathy"),
    ("Community Building", "Role of Education"),
    ("Community Building", "Inclusion"),
    ("Community Building", "Social Responsibility"),
    ("Compassion and Empathy", "Empathy"),
    ("Compassion and Empathy", "Inclusion"),
    ("Inclusion", "Cultural Diversity"),
    ("Inclusion", "Social Responsibility"),
    ("Role of Youth", "Leadership"),
    ("Role of Youth", "Dreams and Aspirations"),
    ("Role of Youth", "Self-love"),
    ("Role of Youth", "Personal Identity"),
    ("Dreams and Aspirations", "Self-love"),
    ("Self-love", "Personal Identity"),
    ("Personal Identity", "Mental Health Awareness"),
    ("Personal Identity", "Overcoming Adversity"),
    ("Overcoming Adversity", "Resilience"),
    ("Resilience", "Environmental Awareness"),
    ("Environmental Awareness", "Global Citizenship"),
("Role of Youth", "Social Responsibility")
]

# Add edges to the graph
G.add_edges_from(edges)

# Define the position of each node using a spring layout
pos = nx.spring_layout(G, seed=42)

# Define colors for different concept categories
node_colors = {
    "Peace": "skyblue",
    "Community Building": "violet",
    "Compassion and Empathy": "orchid",
    "Role of Education": "orange",
    "Inclusion": "deepskyblue",
    "Cultural Diversity": "lightcoral",
    "Social Responsibility": "tomato",
    "Role of Youth": "lightgreen",
    "Leadership": "plum",
    "Dreams and Aspirations": "gold",
    "Self-love": "lightsalmon",
    "Personal Identity": "khaki",
    "Mental Health Awareness": "lightpink",
    "Overcoming Adversity": "lightsteelblue",
    "Resilience": "turquoise",
    "Environmental Awareness": "yellowgreen",
    "Global Citizenship": "peachpuff",
    "Empathy": "lightblue"
}

# Draw the nodes with sizes proportional to their frequencies and colors
node_sizes = [G.nodes[concept].get('size', 100) for concept in G.nodes]
node_colors_list = [node_colors[concept] for concept in G.nodes]
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors_list, alpha=0.7)

# Draw the edges
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)

# Prepare for label adjustments
texts = []
for concept, (x, y) in pos.items():
    texts.append(plt.text(x, y, concept, fontsize=6, color='black', ha='center', va='center'))

# Adjust labels to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle="->", color='black', lw=1))

# Display the graph
plt.title("Concept Mind Map")
plt.axis("off")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add nodes with frequencies
G.add_node("Indian Political System", frequency=34)
G.add_node("Constitution", frequency=6)
G.add_node("Political Theory and Thought", frequency=24)
G.add_node("Indian Government and Politics", frequency=22)
G.add_node("Comparative Politics", frequency=18)

# Add edges with thickness
G.add_edge("Indian Political System", "Constitution", thickness=3)
G.add_edge("Constitution", "Fundamental Rights", thickness=3)
G.add_edge("Constitution", "Directive Principles", thickness=3)
G.add_edge("Political Theory and Thought", "Western Political Thought", thickness=2)
G.add_edge("Political Theory and Thought", "Indian Political Thought", thickness=2)
G.add_edge("Indian Government and Politics", "Political Parties", thickness=2)
G.add_edge("Indian Government and Politics", "Elections", thickness=2)
G.add_edge("Comparative Politics", "Political Systems", thickness=2)

# Draw the graph
nx.draw(G, with_labels=True, node_size=1500, node_color="lightblue", edge_color="gray", linewidths=2, font_size=12)

# Display the graph
plt.show()
```
